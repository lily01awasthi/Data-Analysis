```python
# performing T-test for compairing between two different type of cereals
cereal_1 = df['sugars'][df["mfr"] == "K"]
cereal_2 = df['sugars'][df["mfr"] == "Q"]
ttest_ind(cereal_1,cereal_2,equal_var=False)
print("Since the  pvalue=0.27> 0.01. This is evident that it is likely that the both data are from same distribution.")

# Histogram for two different groups of manufacturers with respect to their sugar content in cereals
plt.hist(cereal_1,label="manufactirer_K")
plt.hist(cereal_2, label="manufactirer_Q")
plt.legend(loc="upper right")
print("From the histogram we can see that the manufacturer Q has cereals with less sugar content than K")
```
